```python
import pandas as pd

corpus = pd.read_csv('../data/corpus_el_google.csv', sep=';')
corpus.head()
```

```python
ref_df = corpus[['text_el', 'google_nmt']]
ref_df.sample(10)
```

```python
def calculate_bleu(target_text, nmt_text):
    """Calculate BLEU score for a reference (human) text and a machine-translated text."""

    from nltk.translate.bleu_score import corpus_bleu

    references = [[sent.split()] for sent in target_text]
    hypotheses = [sent.split() for sent in nmt_text]
    bleu_score = 100 * corpus_bleu(references, hypotheses)

    # interpretation according to https://cloud.google.com/translate/automl/docs/evaluate#interpretation
    if bleu_score < 20:
        interpretation = 'Hard to get the gist'
    elif bleu_score < 40:
        interpretation = 'Understandable to good translations'
    elif bleu_score < 60:
        interpretation = 'High to very high quality translations'
    else:
        interpretation = 'Quality often better than human'

    return f'BLEU score: {bleu_score:.2f} ({interpretation})'
```

```python
bleu_google = calculate_bleu(ref_df['text_el'], ref_df['google_nmt'])
bleu_google
```
